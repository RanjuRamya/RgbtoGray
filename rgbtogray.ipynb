```python
#convert rgb to grayscale using opencv
#cv computer vision


import cv2
```

```python
img=cv2.imread('flower.jpg')
```

```python
type(img)
```

```python
#img shape
img.shape
```

```python
grayscale_image=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape #color channel is 1
```

```python
cv2.imshow('',grayscale_image)
cv2.waitKey(0)

```

```python
cv2.imwrite('foer_gray.jpg',grayscale_image)
```
